# Subword Tokenization

```python
string="I am learning Tokenization"
```

```python
newString=string.split()
```

```python
newString
```

```python
list3=[]
for word in newString:
  if word[-3:]=="ing":
    list3.append(word.partition("ing"))
  elif word[-7:]=="ization":
    list3.append(word.partition("ization"))
  else:
    list3.append(word)
print(list3)
```

```python
list4=[]
for element in list3:
  if type(element) is tuple:
    list4.append(element[0])
    list4.append(element[1])
  else:
    list4.append(element)
print(list4)
```
