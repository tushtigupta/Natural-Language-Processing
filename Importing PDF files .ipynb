```python
f=r"C:\Users\Tushti Gupta\Desktop\sample-local-pdf.pdf"
pdf_text=[]
pdf_reader=PyPDF2.PdfReader(f)
for p in range(len(pdf_reader.pages)):
    pages=pdf_reader.pages[p]
    pdf_text.append(pages.extract_text())
```

```python
pdf_text
```

```python
for page in pdf_text:
    print(page)
    print("\n")
    print("\n")
    print("\n")
```
